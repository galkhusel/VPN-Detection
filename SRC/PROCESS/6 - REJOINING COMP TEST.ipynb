# DECIDING FEATURES AFTER ENCODING.



```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
#['attack_time_x', 'attacker_ip_enum_x', 'label_x', 'attack_time_y',
#       'watcher_country', 'watcher_as_num', 'watcher_as_name',
#       'attacker_country', 'attacker_as_num', 'attacker_as_name',
#       'attack_type', 'watcher_uuid_enum', 'attacker_ip_enum_y', 'label_y',
#       'protocol', 'joined_countries', 'port_count', 'HAS_TCP?', 'HAS_UDP?',
#       'risky_ports', 'attack_time', 'attacker_ip_enum']


BASIC_COLUMNS = ['port_count',  'risky_ports', 'attacker_ip_enum']
OHE_COLUMNS = ['attack_type']
BINARY_COLUMNS = ["joined_countries", 'protocol']
MEAN_COLUMNS = ["attacker_as_name",  "attacker_country"]


DATA_FOLDER = "../../DATA/"
```

# BUILDING


```python
BASIC = pd.read_parquet(DATA_FOLDER + "PROCESSING/PREPEARED_COMP_TEST.parq", columns=BASIC_COLUMNS)
MEAN = pd.read_parquet(DATA_FOLDER + "ENCODED/PCT_MEAN.parq", columns=MEAN_COLUMNS)
df = pd.concat([BASIC, MEAN,], axis=1)
del BASIC
del MEAN

for x in BINARY_COLUMNS:
    aux = pd.read_parquet(DATA_FOLDER + 'ENCODED/COMP_COLUMN_Binary_{}.parq'.format(x))
    print(aux.columns)
    df = pd.concat([df, aux], axis=1)
    del aux

for x in OHE_COLUMNS:
    aux = pd.read_parquet(DATA_FOLDER + 'ENCODED/COMP_COLUMN_OHE_{}.parq'.format(x))
    print(aux.columns)
    df = pd.concat([df, aux], axis=1)
    del aux

df.to_parquet(DATA_FOLDER + 'FINAL/COMP_TEST_FINISHED.parq', index=True)
print(df.columns)
df
```
