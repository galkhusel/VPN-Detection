# DECIDING FEATURES AFTER ENCODING.



```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
#['attack_time_x', 'attacker_ip_enum_x', 'label_x', 'attack_time_y',
#       'watcher_country', 'watcher_as_num', 'watcher_as_name',
#       'attacker_country', 'attacker_as_num', 'attacker_as_name',
#       'attack_type', 'watcher_uuid_enum', 'attacker_ip_enum_y', 'label_y',
#       'protocol', 'joined_countries', 'port_count', 'HAS_TCP?', 'HAS_UDP?',
#       'risky_ports', 'attack_time', 'attacker_ip_enum']


BASIC_COLUMNS = ['port_count', 'HAS_TCP?', 'HAS_UDP?', 'risky_ports', "label"]
OHE_COLUMNS = ['attack_type']
BINARY_COLUMNS = ["joined_countries", 'protocol']
MEAN_COLUMNS = ["attacker_as_name", "watcher_as_name", "attacker_country"]

features_dictionary = {"Binary":BINARY_COLUMNS, "OHE" : OHE_COLUMNS}

DATAFRAMES = ["TRAIN", "TEST", "VAL"]

DATA_FOLDER = "../../DATA/"
```

# BUILDING


```python
for x in DATAFRAMES:
    print(x)
    BASIC = pd.read_parquet(DATA_FOLDER + "/PROCESSING/{}_SPLIT.parq".format(x), columns=BASIC_COLUMNS)
    print(len(BASIC))
    MEAN = pd.read_parquet(DATA_FOLDER + '/ENCODED/{}_MEAN.parq'.format(x), columns=MEAN_COLUMNS)
    print(len(MEAN))
    df = pd.concat([BASIC, MEAN,], axis=1)
    print(len(df))
    del BASIC
    del MEAN
    for y in features_dictionary:
        for z in features_dictionary[y]:
            aux = pd.read_parquet(DATA_FOLDER + '/ENCODED/X_{}_{}_{}.parq'.format(x, y, z))
            print(aux.columns)
            df = pd.concat([df, aux], axis=1)
            del aux
    df.to_parquet(DATA_FOLDER + '/FINAL/{}_FINISHED.parq'.format(x), index=True)
    print(df.columns)
    df
```
