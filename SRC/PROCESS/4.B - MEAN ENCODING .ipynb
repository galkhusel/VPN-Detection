```python
import pandas as pd
```

```python
DATA_FOLDER = "../../DATA/"
train = pd.read_parquet(DATA_FOLDER + "/PROCESSING/TRAIN_SPLIT.parq")
test = pd.read_parquet(DATA_FOLDER + "/PROCESSING/TEST_SPLIT.parq")
val = pd.read_parquet(DATA_FOLDER + "/PROCESSING/VAL_SPLIT.parq")
COMP_TEST = pd.read_parquet(DATA_FOLDER + "/PROCESSING/PREPEARED_COMP_TEST.parq")
```

#MEAN ENCODER


```python
COMP_TEST
```

```python
target_features = [ 'watcher_country',  'watcher_as_name', 'attacker_country', 'attacker_as_name', 'attack_type', 'protocol', 'joined_countries']
for x in target_features:
    train_mean = train.groupby([x])['label'].mean()
    print(train_mean)
    train[x] = train[x].map(train_mean)
    test_mean = test.groupby([x])['label'].mean()
    test[x] = test[x].map(test_mean)
    val_mean = val.groupby([x])['label'].mean()
    val[x] = val[x].map(val_mean)
    COMP_TEST[x] = COMP_TEST[x].map(train_mean).fillna(train_mean.mean())
```

```python
train.to_parquet(DATA_FOLDER + '/ENCODED/TRAIN_MEAN.parq', index=True)
test.to_parquet(DATA_FOLDER + '/ENCODED/TEST_MEAN.parq', index=True)
val.to_parquet(DATA_FOLDER + '/ENCODED/VAL_MEAN.parq', index=True)
COMP_TEST.to_parquet(DATA_FOLDER + '/ENCODED/PCT_MEAN.parq', index=True)
```
